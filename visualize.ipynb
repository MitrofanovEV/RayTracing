```python
import matplotlib.pyplot as plt
import numpy as np
import glob

paths = glob.glob('rays/*')

plt.figure(figsize=(10, 6))
for path in paths:
    try:
        data = np.loadtxt(path) 
        x = data[:, 0]
        z = data[:, 1]
        
        plt.plot(x, z, label=path)
        plt.xlabel('Distance (m)')
        plt.ylabel('Depth (m)')
    except:
        pass

plt.title('Ultrasound Ray Tracing in Water')
plt.gca().invert_yaxis()  
plt.legend()
plt.grid(True)
plt.show()
```
